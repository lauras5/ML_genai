# Exercise: Mean & Median Customer Spend

Generate some random e-commerce data; just an array of total amount spent per transaction. Select the code block, and hit "play" to execute it:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

Now, find the mean and median of this data. In the code block below, write your code, and see if your result makes sense:

```python
print(f"Mean: {np.mean(incomes)}")
print(f"Median: {np.median(incomes)}")
```

Try playing with the code above to generate different distributions of data, or add outliers to it to see their effect.

```python
incomes = np.append(incomes, [50000000])
```

```python
print(f"Outlier Mean: {np.mean(incomes)}")
print(f"Outlier Median: {np.median(incomes)}")
plt.hist(incomes, 50)
plt.show()
```

## Conclusion: Adding outliers changes the mean, but usually not the median

```python
incomes = np.append(incomes, [500000000])
print(f"Outlier Mean: {np.mean(incomes)}")
print(f"Outlier Median: {np.median(incomes)}")
plt.hist(incomes, 50)
plt.show()
```

```python
from scipy import stats
mode_result = stats.mode(incomes)
print(f"Mode: {mode_result.mode}")
```
