# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 50.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```

## Activity

Experiment with different parameters on the normal function, and see what effect it has on the shape of the distribution. How does that new shape relate to the standard deviation and variance?

```python
# NOTE: np.random.normal(loc=0.0, scale=1.0, size=None), (mean, stdev, # of samples)
incomes1 = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes1, 50, alpha=0.5, label='Incomes1')
plt.legend(loc='upper right')
plt.show()

print(f"Incomes1 - Standard Deviation: {incomes1.std()}, Variance: {incomes1.var()}")
```

```python
incomes2 = np.random.normal(150.0, 40.0, 10000)

plt.hist(incomes2, 50, alpha=0.5, label='Incomes2')
plt.legend(loc='upper right')
plt.show()

print(f"Incomes2 - Standard Deviation: {incomes2.std()}, Variance: {incomes2.var()}")
```

### Change sample size, SD and Variance stabalize and you increase the sample size

```python
sample_sizes = [10, 100, 1000, 10000, 50000]
for size in sample_sizes:
    sample = np.random.normal(100.0, 50.0, size)
    print(f"Sample Size: {size}, Standard Deviation: {sample.std():.2f}, Variance: {sample.var():.2f}")
```

```python
mean = incomes.mean()
std_dev = incomes.std()

plt.hist(incomes, 50, alpha=0.6, color='g')
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(mean + std_dev, color='b', linestyle='dashed', linewidth=1) # SD above mean
plt.axvline(mean - std_dev, color='b', linestyle='dashed', linewidth=1) # SD below mean
plt.show()
```

```python
incomes_with_outliers = np.append(incomes, [1000, 2000, 3000])
plt.hist(incomes_with_outliers, 50)
plt.show()

print(f"Incomes with Outliers - Standard Deviation: {incomes_with_outliers.std()}, Variance: {incomes_with_outliers.var()}")
```
