```python
# Average homes in my zip code
# distribution avg = 930000, stddev of 10000
vals = np.random.normal(930000, 10000, 20000)

plt.hist(vals, 100)
plt.show()

print(f"25th Percentile: {np.percentile(vals, 25)}")
print(f"50th Percentile (median): {np.percentile(vals, 50)}")
print(f"75th Percentile: {np.percentile(vals, 75)}")
```

```python
# plotting percentile lines on histogram
plt.hist(vals, 100, alpha=0.6, color='g')
plt.axvline(np.percentile(vals, 25), color='r', linestyle='dashed', linewidth=1, label='25th Percentile')
plt.axvline(np.percentile(vals, 50), color='b', linestyle='dashed', linewidth=1, label='50th Percentile (Median)')
plt.axvline(np.percentile(vals, 75), color='y', linestyle='dashed', linewidth=1, label='75th Percentile')
plt.legend()
plt.show()
```

```python
# calculate different percentiles
percentiles = [10, 25, 50, 75, 90, 99]
for p in percentiles:
    print(f"{p}th Percentile: {np.percentile(vals, p)}")
```

### Cumulative Distribution Function (CDF) 

```python
sorted_vals = np.sort(vals)
cdf = np.arange(1, len(sorted_vals) + 1) / len(sorted_vals)

plt.plot(sorted_vals, cdf)
plt.axhline(0.25, color='r', linestyle='dashed', label='25th Percentile')
plt.axhline(0.5, color='b', linestyle='dashed', label='50th Percentile (Median)')
plt.axhline(0.75, color='y', linestyle='dashed', label='75th Percentile')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

### Calculating Interquartile Range (IQR)

```python
iqr = np.percentile(vals, 75) - np.percentile(vals, 25)
print(f"Interquartile Range (IQR): {iqr}")
```
