# Moments: Mean, Variance, Skew, Kurtosis

Create a roughly normal-distributed random set of data:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

The first moment is the mean; this data should average out to about 0:

```python
np.mean(vals)
```

The second moment is the variance:

```python
np.var(vals)
```

The third moment is skew - since our data is nicely centered around 0, it should be almost 0:

```python
import scipy.stats as sp
sp.skew(vals)
```

The fourth moment is "kurtosis", which describes the shape of the tail. For a normal distribution, this is 0:

```python
sp.kurtosis(vals)
```

Understanding skew: change the normal distribution to be centered around 10 instead of 0, and see what effect that has on the moments.

The skew is still near zero; skew is associated with the shape of the distribution, not its actual offset in X.

```python
# lognormal --> positively skewed
skew_vals = np.random.normal(10, 1, 10000) 

plt.hist(skew_vals, 10)
plt.show()

print("\n")
print(f"First Moment (Mean):       {np.mean(skew_vals)}")
print(f"Second Moment (Variance):  {np.var(skew_vals)}")
print(f"Third Moment (Skew):       {sp.skew(skew_vals)}")
print(f"Fourth Moment (Kurtosis):  {sp.kurtosis(skew_vals)}\n")
```

```python
# lognormal --> positively skewed
skew_vals = np.random.lognormal(0, 1, 10000) 

plt.hist(skew_vals, 10)
plt.show()

print("\n")
print(f"First Moment (Mean):       {np.mean(skew_vals)}")
print(f"Second Moment (Variance):  {np.var(skew_vals)}")
print(f"Third Moment (Skew):       {sp.skew(skew_vals)}")
print(f"Fourth Moment (Kurtosis):  {sp.kurtosis(skew_vals)}\n")
```

```python
# positively skewed
exp_vals = np.exp(vals) 

plt.hist(exp_vals, 10)
plt.show()

print("\n")
print(f"First Moment (Mean):       {np.mean(exp_vals)}")
print(f"Second Moment (Variance):  {np.var(exp_vals)}")
print(f"Third Moment (Skew):       {sp.skew(exp_vals)}")
print(f"Fourth Moment (Kurtosis):  {sp.kurtosis(exp_vals)}\n")
```

```python
from scipy.stats import skewnorm

# positive for right skew, negative for left skew
skewness = 10

# Generate values from a skew-normal distribution
skew_vals = skewnorm.rvs(a=skewness, size=10000)

plt.hist(skew_vals, 50)
plt.show()

print("\n")
print(f"First Moment (Mean):       {np.mean(skew_vals)}")
print(f"Second Moment (Variance):  {np.var(skew_vals)}")
print(f"Third Moment (Skew):       {sp.skew(skew_vals)}")
print(f"Fourth Moment (Kurtosis):  {sp.kurtosis(skew_vals)}\n")
```

```python
# same code but negative skew
skewness = -10  

skew_vals = skewnorm.rvs(a=skewness, size=10000)

plt.hist(skew_vals, 50)
plt.show()

print("\n")
print(f"First Moment (Mean):       {np.mean(skew_vals)}")
print(f"Second Moment (Variance):  {np.var(skew_vals)}")
print(f"Third Moment (Skew):       {sp.skew(skew_vals)}")
print(f"Fourth Moment (Kurtosis):  {sp.kurtosis(skew_vals)}\n")
```
